```python
# selected_assay_list_o = os.listdir("/content/29assays/rf")
import pandas as pd
for m in metric_list : 
  df_index2 = ['val','test',
            'val_tb_auc','test_tb_auc',
            'val_tb_f1','test_tb_f1',
            'val_tb_kappa','test_tb_kappa',
            'val_tb_mcc','test_tb_mcc']
  df_index=[]
  for df_ in df_index2 : 
    df_index.append(m+'_'+df_)
  df = pd.DataFrame(columns =selected_assay_list, index = df_index)
  for index, ss in enumerate(selected_assay_list) : 
    for i,f in enumerate(file_list) : 
      try : 
        result_df = pd.read_csv('/content/29assays_dd/rf/' + ss  +'/'+f +'_' +ss + '.csv')
        try : 
          df.iloc[i,index] = result_df.loc[10,m]
        except : 
          df.iloc[i,index] = result_df.loc[0,m]
      except : 
        print("error_assay : ",'/content/29assays_dd/rf/' + ss  +'/'+f +'_' +ss + '.csv')
  df.to_csv('result_dd/'+m+"_result.csv")
```
